## Few Shot prompts

Few-shot prompting allows us to give a few example interactions to the LLM so that it continues to follow the pattern. This increases the likelyhood of gett the response we want.

We don't want to send the full manuscript to the LLM when we are handling metadata, Story and StoryDialog are stored separately. They map 1:1.

## Entity Relationship Diagram

### Story ERD

```python
import erdantic as erd
from erdantic.core import EntityRelationshipDiagram, Edge, Cardinality, Modality, FullyQualifiedName
from model import Story, StoryDialog, Character, CharacterRelationship, Scene, DialogueLine, SceneDialog

# Create the diagram
diagram = erd.create(Story)

# Customize appearance
edge_attributes = {
    # "rankdir": (("ordering", "out"),)
}

graph_attr = {
    # "rankdir": "TD",
    # "nodesep": "1.0",
    # "ranksep": "1.0",
}

# Save the diagram to a file
diagram.draw("data_model_story.png", graph_attr=graph_attr, edge_attr=edge_attributes)

# Display the ERD
from IPython.display import Image
display(Image("data_model_story.png"))
```
