# Compare Models

**Note**: *Some models have very different parameter counts*

````python
# from model_text import llm, llm_json, display_messages
from llama_index.core.llms import ChatMessage

def creativity_prompt(prompt, system=None):
    if system is None:
        system = (
            "You are a creative story writer. "
            "Respond to user queries with a one line imaginative and engaging story summary. "
        )
    return [
        ChatMessage(role="system", content=system),
        ChatMessage(role="user", content=prompt),
    ]

creativity_prompts = [
    "A one-legged-chicken",
    "A cyberpunk detective",
    "A pen",
    "A cat",
]
````

````python
from llama_index.llms.ollama import Ollama
from IPython.display import Markdown, display

model_names = [
    "nemotron:70b",         # 128k context window - Llama-3.1-Nemotron-70B-Instruct
    "hermes3:70b",          # 128k context window
    # "qwen2.5:72b",          # 128k context window
    "llama3.1:70b",         # 128k context window
    "gemma2:27b",           # 32k context window
    "mistral-openorca:7b",  # 32k context window
    # "phi3.5",             # 128k context window
]
````

````python
# Pull the models
for model_name in model_names:
    !ollama pull {model_name}
````

````python
# Actually run the models
display(Markdown("# LLM Model Creativity Comparison\n\n"))

for model_name in model_names:
    llm = Ollama(model=model_name, temperature=0.0)
    display(Markdown(f"## {model_name}:\n\n"))
    for prompt in creativity_prompts:
        response = llm.chat(creativity_prompt(prompt), max_new_tokens=100, timeout=300)
        display(Markdown("> " + response.message.content.replace("\n", "\n> ")))
````

# JSON Handling

````python
import json

prompts = [
    "Provide a JSON array of three different fruits, each with a name and color. ",
    "Create a JSON object representing a book. It should include the title, author, publication year, and genres (as an array). ",
    "Create a json object that has a content field and in there is a markdown article 3 paragraphs long about cats",
]
system = (
    "You are a JSON encoder. "
    "Do whatever the user says and convert it to a valid json object. "
    "Do not include any additional information. "
    "Do not use markdown. Just use pure JSON."
    "Only respond with the JSON object. "
)

display(Markdown("# LLM Model JSON Comparison\n\n"))

for model_name in model_names:
    llm = Ollama(model=model_name, temperature=0.0, request_timeout=300.0,)
    display(Markdown(f"## {model_name}:\n\n"))
    for prompt in prompts:
        response = llm.chat(creativity_prompt(prompt, system), max_new_tokens=100, timeout=300)
        content = response.message.content
        try:
            content = json.loads(content).dumps(content, indent=2)
        except:
            print("Error parsing JSON")
        display(Markdown(f"### Prompt: {prompt}\n\n"))
        display(Markdown("---\n\n```json\n" + content + "\n```\n\n---\n\n"))
        # print(repr(response.message.content))
````
