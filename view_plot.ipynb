# Diagrams, Beat Sheets and More

There are several types of diagrams and frameworks that storytellers use to plan stories for books, movies, and other narrative-driven media. Here are some of the most popular:

1. **Story Arc or Freytag's Pyramid**
    - **Description**: This diagram is based on the traditional plot structure and is shaped like a pyramid. It includes stages such as exposition, rising action, climax, falling action, and resolution.
    - **Use**: This structure is commonly used to outline the progression of a story and the build-up to the climax, showing how tension increases and resolves.

2. **Three-Act Structure**
    - **Description**: Divided into three parts â€” Setup, Confrontation, and Resolution.
    - **Use**: The three-act structure is popular in screenwriting, guiding writers to introduce the characters and setting, then escalate conflict, and finally resolve the story in the third act.

3. **Mind Maps**
    - **Description**: Mind maps branch out ideas from a central theme, allowing creators to visually explore different aspects of their story.
    - **Use**: Writers often use mind maps to brainstorm characters, settings, themes, and subplots.

4. **Hero's Journey (or Monomyth)**
    - **Description**: A circular diagram that includes stages such as the call to adventure, meeting with a mentor, facing challenges, and returning transformed.
    - **Use**: This structure is especially common in epic tales and fantasy. It provides a guide for stories where a protagonist undergoes a transformative journey.

5. **Beat Sheets**
    - **Description**: Lists of major "beats" or turning points within a story, usually in a linear form.
    - **Use**: Beat sheets are widely used in screenwriting to break down every significant event, helping writers ensure consistent pacing and story flow.

6. **Character Arcs**
    - **Description**: Diagrams or graphs that track a character's development throughout the story, showing emotional highs and lows, changes in personality, and growth.
    - **Use**: Character arcs are essential for character-driven stories, helping authors visualize how each character evolves.

7. **Timeline Diagrams**
    - **Description**: A horizontal line with events plotted in chronological order.
    - **Use**: Particularly useful for complex stories or those with multiple timelines, it helps authors keep track of events and ensure consistency.

8. **Scene Diagrams**
    - **Description**: Scenes are mapped out as boxes connected by lines, showing the order and relationships of events.
    - **Use**: Writers use scene diagrams to plan individual scenes and how they contribute to the story's flow and pacing.


# Three-Act Structure

Not necessarily 3 acts, but here we go.

The **Three-Act Structure** is a classic storytelling framework that divides a narrative into three main sections: **Act 1 (Setup)**, **Act 2 (Confrontation)**, and **Act 3 (Resolution)**. This structure is widely used in screenwriting, novels, and other forms of storytelling. It provides a clear and predictable flow that guides the story's progression, helping writers create narratives with engaging arcs and satisfying resolutions.

### Breakdown of the Three-Act Structure

#### **Act 1: Setup (Beginning)**
   - **Purpose**: Introduce the main characters, setting, and the story's premise. The protagonist's normal life is established, and the central conflict begins to emerge.
   - **Key Elements**:
     - **Exposition**: Introduces the main characters, setting, and tone of the story.
     - **Inciting Incident**: A crucial event that disrupts the protagonist's world and kick-starts the plot. It poses a question or problem that the story will resolve.
     - **First Plot Point**: At the end of Act 1, the protagonist makes a choice or is forced into action, committing them to the story's journey. This event propels them from the known world into the unknown.

#### **Act 2: Confrontation (Middle)**
   - **Purpose**: The protagonist faces challenges and obstacles that test their resolve, often causing inner conflict and growth. This act forms the bulk of the story, raising stakes and creating tension.
   - **Key Elements**:
     - **Rising Action**: A series of events and challenges intensify the central conflict, deepening the stakes and pushing the protagonist towards growth or change.
     - **Midpoint**: The story reaches a pivotal moment, often where the protagonist faces a major success or setback, changing their understanding or approach. The stakes are raised even higher.
     - **Second Plot Point**: Near the end of Act 2, the protagonist faces a crisis or low point, often feeling defeated or questioning their choices. This moment leads directly to the climax.

#### **Act 3: Resolution (End)**
   - **Purpose**: Resolve the story's central conflict, bringing the character's journey to a satisfying conclusion.
   - **Key Elements**:
     - **Climax**: The highest point of tension and conflict, where the protagonist faces the primary antagonist or challenge directly. The outcome of this confrontation determines the story's resolution.
     - **Falling Action**: After the climax, any remaining issues are resolved, and loose ends are tied up.
     - **Denouement**: The story wraps up, showing how the protagonist and their world have changed. The protagonist may return to their normal life with new insights or a transformed perspective.

### Visualizing the Three-Act Structure

A Three-Act Structure can be visualized like this:

```plaintext
        Act 1           Act 2           Act 3
      (Setup)      (Confrontation)   (Resolution)
      
Exposition    Rising Action     Climax      Denouement
    \             /                 \             /
     \_________/                     \_________/
```

### Example of the Three-Act Structure in Practice

Here's an example of how the Three-Act Structure might look in a typical hero's journey:

- **Act 1**: Luke Skywalker lives a quiet life on Tatooine until he discovers a message from Princess Leia, setting him on a path to join the Rebellion.
- **Act 2**: Luke trains with Obi-Wan Kenobi, faces challenges, and suffers losses. At the midpoint, he commits fully to the cause. By the end of Act 2, he's ready to confront the Empire.
- **Act 3**: Luke and the Rebels attack the Death Star in a climactic battle. After a high-stakes confrontation, he successfully destroys it, and the galaxy celebrates the victory.

The Three-Act Structure is a time-tested framework that helps create compelling, well-paced narratives with strong arcs, clear turning points, and meaningful resolutions.

````python
def render_three_act_structure(story):
    """
    Render the acts of a Story object in a 3-act structure format for a writer.
    """
    output = f"# {story.title}\n\n**Genre**: {story.genre}\n\n**Plot Overview**:\n{story.plot_overview}\n\n"
    
    for act in story.acts:
        output += f"## {act.title}\n\n"
        output += f"**Description**: {act.description}\n\n"
        
        for scene in act.scenes:
            output += f"- **Scene: {scene.title}**\n"
            output += f"  * **Description**: {scene.description}\n"
            output += f"  * **Characters Involved**: {', '.join(scene.characters_involved)}\n"
            output += f"  * **Setting**: {scene.setting}\n"
            output += f"  * **Mood**: {scene.mood}\n"
            output += f"  * **Key Actions**:\n"
            for action in scene.key_actions:
                output += f"      - {action}\n"
            output += "\n"
    
    return output
````


# Beat Sheet

A **beat sheet** is a structured outline that breaks down the major plot points, or "beats," in a story. Each beat represents a specific moment or turning point that drives the plot forward or develops character arcs. Beat sheets are especially popular in screenwriting and are often used to map out the pacing, tension, and emotional impact of a story before writing the full script or manuscript.

A beat sheet serves as a roadmap, helping writers ensure their story progresses logically and maintains momentum. It can be adapted to fit various storytelling structures, including the Three-Act Structure, the Hero's Journey, and more.

### Common Beats in a Beat Sheet

While different formats exist, these are the beats typically included in a standard beat sheet for a Three-Act Structure:

#### **Act 1: Setup**
1. **Opening Image**: The first impression of the story, showing the protagonist's world or state before change. Sets the tone and provides a visual hook.
2. **Theme Stated**: A hint or statement about the central theme, often presented subtly within dialogue or the setting.
3. **Setup**: Introduction to the main characters, setting, and the protagonist's ordinary world, establishing what's at stake.
4. **Catalyst (Inciting Incident)**: The event that disrupts the protagonist's life and introduces the central conflict.
5. **Debate**: The protagonist reacts to the catalyst, hesitating or considering options before fully committing to their journey.

#### **Act 2A: Confrontation (First Half of Act 2)**
6. **Break into Act 2**: The protagonist makes a choice or takes action, leaving their ordinary world behind to confront the main conflict.
7. **B Story**: Introduction of a subplot or secondary story (often a love interest or friendship) that adds depth to the main plot.
8. **Fun and Games**: A lighter section where the protagonist explores the new world or situation, often providing a glimpse of what success could look like.
9. **Midpoint**: The story's turning point. The protagonist faces a major event that raises the stakes, often changing the direction of the story or revealing new information.

#### **Act 2B: Confrontation (Second Half of Act 2)**
10. **Bad Guys Close In**: Obstacles intensify as the protagonist faces increasing opposition, and the tension builds. The protagonist may struggle or make mistakes.
11. **All Is Lost**: The protagonist hits rock bottom, facing their darkest moment or greatest setback.
12. **Dark Night of the Soul**: The protagonist reflects on their journey and struggles with despair, fear, or self-doubt, leading to an inner realization or change.

#### **Act 3: Resolution**
13. **Break into Act 3**: The protagonist, renewed with resolve or insight, takes decisive action to confront the central conflict.
14. **Finale**: The climax and resolution of the story. The protagonist faces their biggest challenge, often leading to victory, transformation, or a significant outcome.
15. **Final Image**: The last moment or image of the story, showing how the protagonist or world has changed since the beginning. Provides closure and reinforces the theme.

### Example Beat Sheet (for a Hero's Journey)

| Beat                      | Description                                                                                      |
|---------------------------|--------------------------------------------------------------------------------------------------|
| Opening Image             | Introduces the hero in their ordinary world                                                     |
| Theme Stated              | A hint of the story's theme (e.g., "You can't run from your destiny.")                           |

| Debate                    | The hero hesitates, unsure whether to accept the challenge                                       |
| Break into Act 2          | The hero commits to the journey, stepping into the unknown                                      |
| Fun and Games             | The hero learns about the new world and begins to grow                                          |
| Midpoint                  | A major event that raises stakes and shifts the hero's goal                                     |
| Bad Guys Close In         | Increasing obstacles and enemies put pressure on the hero                                       |
| All Is Lost               | The hero faces their greatest failure or setback                                                |
| Dark Night of the Soul    | The hero reflects, finding renewed strength or insight                                          |
| Break into Act 3          | The hero takes decisive action to confront the main conflict                                    |
| Finale                    | The hero faces the ultimate test, leading to transformation and resolution                      |
| Final Image               | Shows the hero's return or newfound strength, providing closure                                 |

### Purpose of a Beat Sheet

A beat sheet helps writers:
- **Organize Story Structure**: It ensures key moments are included and that the story flows logically from one beat to the next.
- **Maintain Pacing and Tension**: By outlining the sequence and timing of events, it keeps the story engaging and avoids pacing issues.
- **Develop Character Arcs**: It maps the protagonist's journey and emotional growth, ensuring that each event contributes to their transformation.
- **Guide the Writing Process**: A beat sheet serves as a blueprint for the draft, helping writers stay focused and avoid plot holes.

Beat sheets are flexible and can be as detailed or minimal as needed. They provide a concise overview of the story's structure, making it easier to maintain focus, balance, and momentum throughout the writing process.

````python
def create_beat_sheet(story: Story) -> str:
    """
    Createds a beat sheet for the given story in a structured Markdown format.
    
    Parameters:
        story (Story): The story object containing acts, scenes, and story beats.
    """
    beat_sheet_md = f"# Beat Sheet for '{story.title}'\n\n"
    beat_sheet_md += f"**Genre**: {story.genre}\n\n"
    beat_sheet_md += f"**Plot Overview**: {story.plot_overview}\n\n"
    
    # Display story beats
    beat_sheet_md += "## Story Beats\n"
    if story.story_beats:
        for beat in story.story_beats:
            beat_sheet_md += f"### {beat.name}\n"
            beat_sheet_md += f"{beat.description}\n"
            if beat.scene:
                beat_sheet_md += f"**Scene Link**: {beat.scene}\n"
            beat_sheet_md += "\n"
    else:
        beat_sheet_md += "_No story beats defined._\n\n"
    
    # Display acts and scenes
    beat_sheet_md += "## Acts and Scenes\n"
    if story.acts:
        for act in story.acts:
            beat_sheet_md += f"### {act.title}\n"
            beat_sheet_md += f"{act.description}\n\n"
            for scene in act.scenes:
                beat_sheet_md += f"#### Scene: {scene.title}\n"
                beat_sheet_md += f"- **Description**: {scene.description}\n"
                beat_sheet_md += f"- **Location**: {scene.location}, **Time of Day**: {scene.time_of_day}\n"
                beat_sheet_md += f"- **Mood**: {scene.mood}\n"
                beat_sheet_md += f"- **Characters Involved**: {', '.join(scene.characters_involved)}\n"
                beat_sheet_md += f"- **Key Actions**: {', '.join(scene.key_actions)}\n"
                beat_sheet_md += "\n"
    else:
        beat_sheet_md += "_No acts and scenes defined._\n"
    
    # Display emotional arc, if available
    beat_sheet_md += "## Emotional Arc\n"
    if story.emotional_arc:
        for stage in story.emotional_arc:
            beat_sheet_md += f"### {stage.stage}\n"
            beat_sheet_md += f"{stage.description}\n\n"
    else:
        beat_sheet_md += "_No emotional arc defined._\n"

    return beat_sheet_md
````


# Character Arc of Protagonists

````python
from graphviz import Digraph
import os

def create_character_arc(character, story_dir):
    # Ensure the directory exists
    os.makedirs(story_dir, exist_ok=True)
    
    # Initialize a directed graph
    dot = Digraph("Character Arc", format="png")
    
    arc = character.character_arc

    if not arc:
        return None
    
    # Add nodes for initial and final states
    dot.node("initial", f"Initial State\n({arc.initial_state})")
    dot.node("final", f"Final State\n({arc.final_state})")
    
    # Add nodes and edges for key moments
    previous_stage = "initial"
    for i, moment in enumerate(arc.key_moments):
        stage_id = f"stage_{i}"
        dot.node(stage_id, moment)
        dot.edge(previous_stage, stage_id)
        previous_stage = stage_id
    dot.edge(previous_stage, "final")
    
    # Define the file path to save the graph
    file_path = os.path.join(story_dir, "character_arc_graph")
    dot.render(file_path)  # Render the graph to a file
    
    return file_path + ".png"  # Return the full path to the rendered image
````

````python
# Generate graph for character arc
from IPython.display import Image, display, Markdown

draft_dir = f"{settings.STORY_DIR}/draft"

for character in story.characters:
    image_path = create_character_arc(character, draft_dir)
    if not image_path:
        print(f"No character arc defined for {character.name}.")
    else:
        display(Markdown(f"## {character.name}"))
        display(Image(filename=image_path))
````


# Timeline Diagram for a Story

A **Timeline Diagram** for a story is a visualization that maps key events, scenes, and character arcs across the chronological timeline of the story. This type of diagram is especially useful for tracking events, understanding temporal relationships between scenes, and managing complex narratives involving multiple plotlines or time jumps.

Here's a guide to building a **Timeline Diagram** for a story, along with a Python function that uses Graphviz to create the diagram.

### Purpose of a Timeline Diagram

- **Chronological Overview**: Shows the order of key events, allowing writers and readers to see how the plot unfolds over time.
- **Tracking Character Arcs**: Helps in tracking character development and involvement across the timeline.
- **Visualizing Parallel Plotlines**: Useful for stories with multiple intersecting or parallel storylines.
- **Managing Time Jumps**: Clarifies stories with flashbacks, flashforwards, or non-linear timelines.
- **Planning and Consistency**: Ensures events and character appearances are consistent and logically ordered.

### Structure of the Timeline Diagram

In this timeline:
- Each **Act** can represent a section along the timeline, with scenes shown sequentially within each act.
- **Events** within scenes are represented as nodes on the timeline, linked in sequence.
- **Characters** involved in each scene can be indicated to show character presence across the timeline.

````python
from graphviz import Digraph
import os
from IPython.display import display, Image, Markdown

def create_timeline_diagram(story: Story, story_dir: str) -> str:
    """
    Creates a Timeline Diagram for the story, with each act flowing vertically
    (top-to-bottom), and scenes within each act flowing horizontally (left-to-right).

    Parameters:
    - story: Story object containing acts and scenes.
    - story_dir: Directory to save the diagram image.

    Returns:
    - Path to the generated diagram image.
    """
    # Ensure the directory exists
    os.makedirs(story_dir, exist_ok=True)
    
    # Initialize a directed graph with left-to-right orientation
    dot = Digraph("Act-Scene Timeline Diagram", format="png")
    dot.attr(label=story.title, fontsize="20", labelloc="top", rankdir="LR")  # Overall left-to-right for acts

    # Add nodes and edges for each act and scene
    for act in story.acts:
        with dot.subgraph(name=f"cluster_{act.act_id}") as sub:
            sub.attr(label=act.title or f"Act {act.act_id}", fontsize="16", style="filled", color="lightgrey", rankdir="TB")

            # Add scenes as timeline nodes within each act, flowing horizontally
            previous_scene_id = None
            for scene in act.scenes:
                scene_label = f"{scene.title}\nTime: {scene.time_of_day}\nLocation: {scene.location}\nCharacters: {', '.join(scene.characters_involved)}"
                sub.node(scene.scene_id, scene_label, shape="ellipse", style="filled", color="lightblue", fontsize="10")
                
                # Link scenes in order within the act
                if previous_scene_id:
                    sub.edge(previous_scene_id, scene.scene_id)
                previous_scene_id = scene.scene_id

    # Link the last scene of each act to the first scene of the next act to show progression
    for i in range(len(story.acts) - 1):
        last_scene_current_act = story.acts[i].scenes[-1].scene_id
        first_scene_next_act = story.acts[i + 1].scenes[0].scene_id
        dot.edge(last_scene_current_act, first_scene_next_act, style="dashed")

    # Define the file path to save the graph
    file_path = os.path.join(story_dir, "act_scene_timeline_diagram")
    dot.render(file_path)  # Render the graph to a file
    
    return file_path + ".png"  # Return the full path to the rendered image
````

````python
# Generate a Graphviz diagram based on the structure
import settings

draft_dir = f"{settings.STORY_DIR}/draft"

scene_diagram_path = create_timeline_diagram(story, draft_dir)

# Display the generated image
display(Markdown(f"## Timeline Diagram for '{story.title}'"))
display(Image(filename=scene_diagram_path))
````


# Scene Diagram for a Screenplay

A **screenplay scene diagram** is a visual tool used by writers, directors, and production teams to map out the structure, flow, and connections between scenes in a screenplay. Here are some of the main purposes and benefits of using a screenplay scene diagram:

### 1. **Visualizing Story Structure**
   - It provides a **bird's-eye view** of the screenplay, allowing writers and directors to see the flow of scenes within each act.
   - By visualizing scenes, it becomes easier to identify structural elements like the **beginning, middle, and end** of the story, as well as to pinpoint pivotal scenes like the **inciting incident, climax, and resolution**.

### 2. **Tracking Character Arcs and Involvement**
   - A scene diagram can show which **characters are involved** in each scene, making it easier to track character arcs and ensure that each character is meaningfully contributing to the plot.
   - It helps identify **gaps or inconsistencies** in character presence and development throughout the screenplay.

### 3. **Ensuring Narrative Continuity**
   - By visualizing scenes in order, a diagram can highlight **narrative gaps, repetitive elements**, or **sudden shifts** that might disrupt the story's flow.
   - This allows writers to see if the **story progression** is logical and smooth, or if there are **missing scenes** or **transitions** that need attention.

### 4. **Assisting with Pacing and Tension**
   - The diagram can reveal the **pacing** of the screenplay by showing where scenes with high tension or emotional impact are placed.
   - It helps ensure that **rising and falling actions** are balanced and appropriately spaced to maintain the audience's engagement.

### 5. **Aiding Collaboration Across Production Teams**
   - For directors, producers, and other members of the production team, a screenplay scene diagram serves as a quick **reference for the screenplay's structure**.
   - The diagram helps team members discuss scenes, transitions, and structural elements **without needing to comb through the entire script** each time.

### 6. **Facilitating Revisions and Edits**
   - The diagram makes it easy to see where scenes could be **added, removed, or shifted** within the structure.
   - It provides an overview of the screenplay that can help during **script editing and revisions**, especially when working on **restructuring the story**.

### 7. **Organizing Production Scheduling**
   - Knowing the order and connection of scenes aids in **organizing the shooting schedule**, as production teams can better understand how scenes relate and determine if certain scenes need to be shot in sequence or if they can be filmed independently.

### Summary of Use Cases

A screenplay scene diagram is not only a writing tool but also a **production and planning asset**. It helps clarify the screenplay's structure, track character involvement, manage pacing, facilitate collaboration, and streamline the editing and production processes.

````python
from graphviz import Digraph
import os
from IPython.display import display, Image, Markdown

def create_screenplay_scene_diagram(story: Story, story_dir: str) -> str:
    """
    Creates a Graphviz diagram of a screenplay's structure by organizing scenes within acts.

    Parameters:
    - story: Story object containing acts and scenes.
    - story_dir: Directory to save the diagram image.

    Returns:
    - Path to the generated diagram image.
    """
    # Ensure the directory exists
    os.makedirs(story_dir, exist_ok=True)

    # Initialize a directed graph
    dot = Digraph("Screenplay Scene Diagram", format="png")
    dot.attr(label=story.title, fontsize="20", labelloc="top", rankdir="TB")  # Top-to-bottom flow for acts

    # Process each act and its scenes
    for act in story.acts:
        with dot.subgraph(name=f"cluster_{act.act_id}") as sub:
            sub.attr(label=act.title or f"Act {act.act_id}", fontsize="16", style="filled", color="lightgrey")

            # Add scenes within the act
            for scene in act.scenes:
                scene_label = f"{scene.title}\nLocation: {scene.location}\nCharacters: {', '.join(scene.characters_involved)}\nDescription: {scene.description}"
                sub.node(scene.scene_id, scene_label, shape="box", style="rounded,filled", color="lightblue", fontsize="10")

            # Link scenes in sequence within the act
            for i in range(len(act.scenes) - 1):
                sub.edge(act.scenes[i].scene_id, act.scenes[i + 1].scene_id)

    # Link acts in sequence, connecting the last scene of one act to the first scene of the next
    for i in range(len(story.acts) - 1):
        last_scene_current_act = story.acts[i].scenes[-1].scene_id
        first_scene_next_act = story.acts[i + 1].scenes[0].scene_id
        dot.edge(last_scene_current_act, first_scene_next_act, style="dashed")

    # Define the file path to save the graph
    file_path = os.path.join(story_dir, f"scene_diagram")
    dot.render(file_path)  # Render the graph to a file
    
    return file_path + ".png"  # Return the full path to the rendered image
````

````python
# Generate a Graphviz diagram based on the structure
import settings

draft_dir = f"{settings.STORY_DIR}/draft"

scene_diagram_path = create_screenplay_scene_diagram(story, draft_dir)

# Display the generated image
display(Markdown(f"## Screenplay Scene Diagram for '{story.title}'"))
display(Image(filename=scene_diagram_path))
````


# Freytag's Pyramid

**Freytag's Pyramid** is a framework for structuring dramatic stories, developed by 19th-century German playwright Gustav Freytag. It organizes a story into five key parts, visualized as a pyramid-shaped arc that traces the rise, climax, and resolution of the plot. This structure helps in analyzing and creating stories with a clear beginning, middle, and end, making it particularly useful for traditional narratives. 

Here are the five stages of Freytag's Pyramid:

1. **Exposition**: This is the story's introduction. It establishes the main characters, setting, and background information, giving the audience a sense of the ordinary world the protagonist inhabits before the central conflict begins.

2. **Rising Action**: This is where the main conflict is introduced, and tension builds as the protagonist faces challenges, obstacles, and smaller conflicts. These events create suspense and develop the plot, leading up to the climax.

3. **Climax**: The story's turning point and the moment of highest tension. Here, the main conflict reaches its peak, and the protagonist faces a decisive challenge or choice. This moment often determines the outcome of the story.

4. **Falling Action**: After the climax, the story moves into the falling action. The conflicts start to resolve, and the story's tension decreases as the protagonist deals with the consequences of their actions.

5. **Denouement (Resolution)**: This is the story's conclusion, where loose ends are tied up, and the protagonist's journey concludes. It provides a sense of closure, showing how the characters and world have changed.

### Visualization of Freytag's Pyramid

Visually, Freytag's Pyramid resembles a triangle or pyramid with the following progression:

```
                                  Climax
                                /       \
                Rising Action               Falling Action
            /                                               \ 
Exposition                                                    Denouement
```

### Example in a Story

In a classic story, Freytag's Pyramid might look like this:

- **Exposition**: A young farm boy, Luke, lives on a distant planet, dreaming of adventure.
- **Rising Action**: Luke joins a rebellion and trains with a mentor to confront a tyrannical empire.
- **Climax**: Luke faces the Empire's leader, Darth Vader, and learns shocking truths about his family.
- **Falling Action**: Luke's allies strike against the Empire, gaining the upper hand.
- **Denouement**: Peace is temporarily restored, and Luke continues his journey with new purpose and allies.

Freytag's Pyramid is a helpful framework for structuring stories with a linear, dramatic arc and is widely used in traditional storytelling forms like plays, novels, and films.


````python
from graphviz import Digraph
import os

def create_freytags_pyramid_graphviz(exposition: str, rising_action: str, climax: str, falling_action: str, denouement: str, story_dir: str) -> str:
    # Ensure the directory exists
    os.makedirs(story_dir, exist_ok=True)
    
    # Initialize a directed graph
    dot = Digraph("Freytag's Pyramid", format="png")
    dot.attr(rankdir='BT')  # Bottom-to-top for a pyramid-like effect
    
    # Define subgraph clusters to control the ranking
    with dot.subgraph() as s:
        s.attr(rank='same')
        s.node("exposition", f"Exposition\n({exposition})")
        s.node("denouement", f"Denouement\n({denouement})")
    
    with dot.subgraph() as s:
        s.attr(rank='same')
        s.node("rising_action", f"Rising Action\n({rising_action})")
        s.node("falling_action", f"Falling Action\n({falling_action})")
    
    # Climax at the top
    dot.node("climax", f"Climax\n({climax})")

    # Connect nodes to form a pyramid shape
    dot.edge("exposition", "rising_action")
    dot.edge("rising_action", "climax")
    dot.edge("climax", "falling_action")
    dot.edge("falling_action", "denouement")
    
    # Define the file path to save the graph
    file_path = os.path.join(story_dir, "freytags_pyramid")
    dot.render(file_path)  # Render the graph to a file
    
    return file_path + ".png"  # Return the full path to the rendered image
````

````python
# Function to generate Freytag's Pyramid using LLM and Graphviz

from graphviz import Digraph
from llama_index.core.llms import ChatMessage
import json
import os
from model import Story
from model_text import llm_json, display_messages
from utils import deindent
from IPython.display import display, Image, Markdown

def generate_freytags_pyramid(story: Story) -> dict:
    # Update system message to prompt for Freytag's Pyramid structure
    system_message = deindent("""
        You are a literary assistant with expertise in story structure. Your task is to organize a story into Freytag's Pyramid, a five-part structure commonly used in storytelling.

        Freytag's Pyramid consists of the following five stages:

        - **Exposition**: The introduction, setting up the main characters, the setting, and the initial situation or background context for the story. This provides necessary information to understand the story world.
        - **Rising Action**: The sequence of events that create tension and develop the central conflict. This section builds suspense, introduces challenges, and shows character motivations or transformations leading to the storyâ€™s climax.
        - **Climax**: The turning point of the story, where the main conflict reaches its peak intensity. This moment often involves a critical decision, revelation, or action by the protagonist that determines the storyâ€™s direction.
        - **Falling Action**: The events following the climax, which begin to resolve the conflict. Tension decreases, and consequences of the climax are explored as the story heads toward resolution.
        - **Denouement**: The resolution or conclusion, where the story's threads are tied up. This final stage reveals the end state of the protagonist and the story world, often reflecting on themes or lessons.

        Your response should be formatted as a JSON object that includes these exact five keys, with a brief summary for each stage:
        
        ```json
        {
            "exposition": "A concise introduction to the setting, main characters, and initial situation.",
            "rising_action": "A summary of key events that increase tension and develop the story's central conflict.",
            "climax": "A short description of the turning point where the main conflict reaches its peak intensity.",
            "falling_action": "A description of the events that move the story toward resolution, reducing tension.",
            "denouement": "A brief conclusion that describes the final state of the protagonist and the world."
        }
        ```

        Important:
        - ONLY respond with the JSON object, containing exactly these five keys.
        - Ensure each stage summary is focused, concise, and only includes essential information.
        - Do not include introductory text, explanations, or extra story content outside the five stages.

        **Examples**:
        Below are examples of stories broken down into Freytag's Pyramid structure. Use these to guide your response.
    """)

    # Construct the message for the LLM with few-shot examples
    messages = [ChatMessage(role="system", content=system_message)]
    exclude_keys = ["character"]
    
    # Append each example to the message history
    for example_story, example_pyramid in few_shot_examples:
        messages.append(ChatMessage(role="user", content=example_story.model_dump_json(indent=4, exclude=exclude_keys)))
        messages.append(ChatMessage(role="assistant", content=json.dumps(example_pyramid, indent=4)))

    # Add the user's story prompt to generate the plot overview
    messages.append(ChatMessage(role="user", content=story.model_dump_json(indent=4, exclude=exclude_keys)))

    display_messages(messages)

    # Request the LLM to generate the plot structure
    response = llm_json.chat(messages, max_new_tokens=500)
    response_json = response.message.content

    if settings.DEBUG:
        display(Markdown(f"### Freytag's Pyramid Response:\n```json\n{response_json}\n```"))

    try:
        # Parse the response JSON for Freytag's Pyramid elements
        pyramid_structure = json.loads(response_json)
        return pyramid_structure

    except json.JSONDecodeError as e:
        display(Markdown(f"### Error decoding JSON: {e}"))
        display(Markdown(f"```\n{response_json}\n```"))
        raise
````
