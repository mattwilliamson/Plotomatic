```python
# Scene Generator

from IPython.display import Markdown, display
from model_text import display_messages
from llama_index.core.llms import ChatMessage
from model_text import llm
from utils import blank_story, deindent

request_excludes = {
    "characters": {
        "__all__": {
            "animation_description": True,
            "voice_description": True,
            "internal_conflict": True,
            "character_arc": True,
            "voice_sample": True,
            "catch_phrase": True,
            "props": True,
            "image_prompt": True,
            "image_prompt_short": True,
        }
    }, 
    "acts": True, 

    "story_beats": True, 
    "subplots":True, 
    "emotional_arc": True, 
    "conflict_type": True, 
    "motifs": True,
    "narrative_perspective": True, 
    "themes": True,
    "props": True,
}

response_excludes = {
    "characters": True,
    "prompt": True,
    "title": True,
    "visual_style": True,
    "time_period": True,
    "location": True,
    "genre": True,
    "medium": True,
    "plot_overview": True,
    "props": True,  
    "acts": {
        "__all__": {
            "scenes": {
                "__all__": {
                    "scene_image_prompt": True,
                    "scene_image_prompt_short": True,
                    "background_animation": True,
                    "background_image_prompt": True,
                    "dialogue": True,
                }
            }
        }
    },
}

def generate_scenes(story: Story) -> Story:
    system_message = deindent("""
        You are a creative story outliner. You will take the rough outline of a story and create the acts and scenes, following industry-standard structure for books and movies..
        Your task is to generate a complete list of acts and scenes based on the provided story, ensuring the story has a well-paced progression with buildup, climax, and resolution.
                              
        **Structure Guidelines:**
            - Follow a three-act or five-act structure, depending on the story's genre and length, to ensure a strong narrative arc:
                - **Act 1** (Setup): Introduce characters, setting, and central conflict. Set the stakes and establish the main story question.
                - **Act 2** (Confrontation): Build tension and develop character arcs. This act should include a midpoint shift where the stakes escalate.
                - **Act 3** (Climax and Resolution): Lead to the story's climax, followed by a resolution that addresses the story's central question and wraps up character arcs.
                - **Additional Acts**: For genres like Greek tragedy or epic storytelling, add Acts 4 and 5, where Act 4 intensifies the conflict and Act 5 brings a dramatic or tragic resolution.
            
            - Each act should contain a title and a list of scenes that progressively build upon the storyline.

        Add a list of acts to the story, where each act contains a title and a list of scenes that describe the progression of the story.
        Each scene should be an object that contains a title, description, time_of_day, location, lighting, mood, setting, key actions, and a list of characters involved.
        Each scene needs a unique scene_id to be used as a foreign key reference.
        Each scene should only be one location. If a scene takes place in multiple locations, split it into multiple scenes.
        The number of scenes should reflect the genre and medium of the story. For example a greek tragedy might have 5 acts.
        Be sure to create enough acts and scenes to cover the key moments and character arcs provided in the story.
                              
        - Each scene should contribute to the progression of the story, with moments that build tension, develop character motivations, or reveal critical information.
        - Ensure the number of scenes reflects the genre and medium, with enough variety and depth to cover key story beats and character arcs.

        Flesh out scenes and acts with details that enhance the narrative and build tension.
        **Do not include camera movements.**
        Select characters only from the provided character list by nickname all lowercase.
        Every act must have a list of scenes which are objects.
        Return only a valid JSON object.
    """)

    messages = [ChatMessage(role="system", content=system_message)]
    # few_shot_examples_and_blank = [blank_story] + few_shot_examples

    for example in few_shot_examples[0:2]:
        messages.append(ChatMessage(role="user", content=example.model_dump_json(exclude=request_excludes)))
        messages.append(ChatMessage(role="assistant", content=example.model_dump_json(include=["acts"], exclude=response_excludes)))

    # Add the user input (the story prompt)
    messages.append(ChatMessage(role="user", content=story.model_dump_json(exclude=request_excludes)))
    
    display_messages(messages)

    # Call the LLM to generate the complete story with scenes
    response = llm_json.chat(messages, max_new_tokens=100000).message.content

    # resp = llm.stream_chat(messages, max_new_tokens=10000)

    # # Print it out as it comes in
    # for r in resp:
    #     print(r.delta, end="")

    # response = r.message.content

    try:
        story_with_scenes = Story.model_validate_json(response)
        # print("story_with_scenes: ", story_with_scenes.model_dump_json(indent=4))
    except:
        print("Exception occurred while parsing the response.")
        print(response)
        raise
    
    return story_with_scenes
```
